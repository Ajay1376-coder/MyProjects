```python
from sklearn.cluster import KMeans
```

```python
X=data
X
```

```python
ke=KMeans(n_clusters=5,random_state=11)
```

```python
model1=ke.fit(X)
```

```python
model1.predict(X)
```

```python
model1.inertia_
```

```python
wcss=[]
for i in range(1,13):
    ke=KMeans(n_clusters=i)
    model1=ke.fit(X)   
    model1.predict(X)
    wcss.append(model1.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,13),wcss)
```

```python
### It seems like elbow is at 2
```

```python
ke=KMeans(n_clusters=2)
model1=ke.fit(X)   
pred=model1.predict(X)
model1.inertia_
```

```python
model1.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
```

```python
print("Accuracy of Model with 2 cluster is ",round(silhouette_score(X,ke.labels_),2))
```

```python
pred
```

```python
e=pd.DataFrame({'Cluster':pred})
```

```python
data_new=pd.concat([e,data],axis=1)
```
